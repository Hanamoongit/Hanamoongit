#### Erweiterte Fehleranalyse
welche die drei Hierarchie-Ebenen
a) Bundesland,
b) Bundesland + Landkreis und
c) Bundesland + Landkreis + Stadt
jeweils pro Jahr und Krankenhaus prüft – und die Prüfungen b) und c) nur ausführt,
wenn auf der jeweiligen Vorstufe tatsächlich Mehrfacheinträge gefunden wurden


Wie es funktioniert:

    Dynamische Spaltenfindung
    Ich suche automatisch nach den tatsächlichen Spaltennamen für Jahr und Krankenhaus,
    damit kein KeyError mehr passiert.

    check_duplicates-Funktion
    Diese kapselt die Logik, um

        anhand einer Liste von Gruppierungsspalten (group_cols)

        alle doppelten Zeilen mit df.duplicated(...) zu finden,

        sie zu sortieren und auszugeben,

        und schließlich das Duplikat-DataFrame zurückzugeben.

    Hierarchische Prüfung

        a) prüft, ob ein Krankenhaus innerhalb eines Bundeslandes im selben Jahr mehrmals gelistet ist.

        Wenn a) Duplikate liefert, wird b) aktiviert und prüft Bundesland + Landkreis.

        Nur wenn auch b) Duplikate hat, folgt c) mit Bundesland + Landkreis + Stadt.
        

So vermeide ich unnötige Checks auf tieferen Ebenen, wenn oben schon alles sauber ist.

```python
#-------------------------------------------------------------------------------------------------
# Spalten bestimmen
#-------------------------------------------------------------------------------------------------
year_col = 'Jahr' if 'Jahr' in df.columns else 'year'
hospital_col = 'Name des Krankenhauses' if 'Name des Krankenhauses' in df.columns else 'Hospital'
hospitalid_col = 'Krankenhaus-ID' if 'Krankenhaus-ID' in df.columns else 'Hospital ID'
standortid_col = 'Standort-ID' if 'Standort-ID' in df.columns else 'Standort ID'
```

```python
#-------------------------------------------------------------------------------------------------
# A) Duplikate pro Jahr und Krankenhaus prüfen ---
#    Jede Kombination BG + LK + Stadt + Krankenhaus darf pro Jahr nur einmal vorkommen
#-------------------------------------------------------------------------------------------------

dup_cols = ['Bundesland', 'Landkreis', 'Stadt', hospital_col, year_col]
dups = df[df.duplicated(subset=dup_cols, keep=False)]

if not dups.empty:
    print(f"\n➤ Gefundene Mehrfach-Einträge ({len(dups)} Zeilen):")
    display(dups.sort_values(dup_cols))
else:
    print("\n✅ Keine doppelten Kombinationen von Bundesland+Landkreis+Stadt+Krankenhaus+Jahr gefunden.")
```

```python
#-------------------------------------------------------------------------------------------------
# B) Städte in mehreren BL/LK-Kombinationen über alle Jahre (unverändert) ---
#-------------------------------------------------------------------------------------------------

df['BL_LK_Kombi'] = df['Bundesland'].astype(str) + " | " + df['Landkreis'].astype(str)

combo_counts = (
    df
    .drop_duplicates(subset=['Stadt', 'BL_LK_Kombi'])
    .groupby('Stadt')['BL_LK_Kombi']
    .nunique()
    .reset_index(name='Anzahl_Kombinationen')
)
multi_kombi = combo_counts[combo_counts['Anzahl_Kombinationen'] > 1]

if not multi_kombi.empty:
    print(f"\n➤ Städte in mehreren Bundesland/Landkreis-Kombinationen ({len(multi_kombi)}):")
    display(multi_kombi.sort_values('Anzahl_Kombinationen', ascending=False))
else:
    print("\n✅ Alle Städte sind jeweils genau einer BL/LK-Kombination zugeordnet.")
```

```python
#-------------------------------------------------------------------------------------------------
# C) Krankenhäuser in mehreren BL/LK-Kombinationen prüfen ---
#    Damit sehe ich, ob ein Krankenhaus im Laufe der Jahre
#    mal einem anderen Landkreis/BL zugeordnet wurde.
#-------------------------------------------------------------------------------------------------
hosp_combo = (
    df
    .drop_duplicates(subset=[hospital_col, 'BL_LK_Kombi'])
    .groupby(hospital_col)['BL_LK_Kombi']
    .nunique()
    .reset_index(name='Anzahl_Kombinationen')
)
multi_hosp = hosp_combo[hosp_combo['Anzahl_Kombinationen'] > 1]

if not multi_hosp.empty:
    print(f"\n➤ Krankenhäuser in mehreren BG/LK-Kombinationen ({len(multi_hosp)}):")
    display(multi_hosp.sort_values('Anzahl_Kombinationen', ascending=False))
else:
    print("\n✅ Alle Krankenhäuser sind stabil einer BL/LK-Kombination zugeordnet.")
```

### weitere Analysen


```python
#-------------------------------------------------------------------------------------------------
# Funktion zur Prüfung und Ausgabe
#-------------------------------------------------------------------------------------------------
def check_duplicates(df, group_cols, label):
    """Gibt alle doppelten Einträge zurück und druckt eine Überschrift."""
    dups = df[df.duplicated(subset=group_cols, keep=False)]
    if not dups.empty:
        print(f"\n🔴 Duplikate auf Ebene {label} ({len(dups)} Zeilen):")
        display(dups.sort_values(group_cols))
    else:
        print(f"\n✅ Keine Duplikate auf Ebene {label}.")
    return dups

#-------------------------------------------------------------------------------------------------
# --- a) Bundesland + Krankenhaus + Jahr ---
#-------------------------------------------------------------------------------------------------
cols_a = ['Bundesland', hospital_col, year_col]
dups_a = check_duplicates(df, cols_a, "a) BL + Krankenhaus + Jahr")

if not dups_a.empty:
    #---------------------------------------------------------------------------------------------
    # --- b) Bundesland + Landkreis + Krankenhaus + Jahr ---
    #---------------------------------------------------------------------------------------------
    cols_b = ['Bundesland', 'Landkreis', hospital_col, year_col]
    dups_b = check_duplicates(df, cols_b, "b) BL + LK + Krankenhaus + Jahr")

    if not dups_b.empty:
        #-----------------------------------------------------------------------------------------
        # --- c) Bundesland + Landkreis + Stadt + Krankenhaus + Jahr ---
        #-----------------------------------------------------------------------------------------
        cols_c = ['Bundesland', 'Landkreis', 'Stadt', hospital_col, year_col]
        _ = check_duplicates(df, cols_c, "c) BL + LK + Stadt + Krankenhaus + Jahr")
    else:
        print("\n→ Ebene c übersprungen, da keine Mehrfacheinträge auf Ebene b gefunden wurden.")
else:
    print("\n→ Ebenen b und c übersprungen, da auf Ebene a keine Mehrfacheinträge vorliegen.")
```

```python

print(f"Verwende Jahr-Spalte: {year_col!r}, Krankenhaus-Spalte: {hospital_col!r}\n")
#-------------------------------------------------------------------------------------------------
# Gewünschte Spalten definieren
#-------------------------------------------------------------------------------------------------
selected_cols = ['Bundesland', 'Stadt', hospital_col, year_col]


# Hilfsfunktion für die Ausgabe
def show_filtered_records(names, title):
    print(f"\n=== {title} ===")
    for name in names:
        print(f"\n→ Krankenhaus: {name}")
        display(
            df[df[hospital_col] == name]
            .sort_values([ 'Bundesland', 'Landkreis', 'Stadt', year_col ])
        )

def show_filtered_records(names, title):
    print(f"\n=== {title} ===")
    for name in names:
        print(f"\n→ Krankenhaus: {name}")
        # hier wählst du nur die gewünschten Spalten aus
        display(
            df[df[hospital_col] == name]
              .sort_values(selected_cols)
              .loc[:, selected_cols]
        )

#-------------------------------------------------------------------------------------------------
# a) die ersten drei Krankenhäuser, die in mehreren Jahren auftauchen
# --------- gleiche BG+LK+Stadt+Krankenhaus, nur Jahr unterschiedlich ---
#-------------------------------------------------------------------------------------------------
group_cols_a = ['Bundesland', 'Landkreis', 'Stadt', hospital_col]
dup_groups = (
    df
    .groupby(group_cols_a)[year_col]
    .nunique()
    .reset_index(name='years_count')
)
hospitals_a = dup_groups[dup_groups['years_count'] > 1][hospital_col].unique()[:3]

show_filtered_records(hospitals_a,
    "a) Top-3 Krankenhäuser mit identischer Orts-Kombi, aber in mehreren Jahren")
```

##### b) mit unterschiedlicher BG/LK/Stadt-Kombination
     

```python
#-------------------------------------------------------------------------------------------------
# --- b) gleiches Krankenhaus, aber unterschiedliche BG/LK/Stadt-Kombination ---
# Erzeuge eine Hilfsspalte für die Ort-Kombi
#-------------------------------------------------------------------------------------------------

df['ort_kombi'] = df['Bundesland'].astype(str) + ' | ' + df['Landkreis'].astype(str) + ' | ' + df['Stadt'].astype(str)

# count unique Ort-Kombis je Krankenhaus
loc_counts = (
    df
    .drop_duplicates(subset=[hospital_col, 'ort_kombi'])
    .groupby(hospital_col)['ort_kombi']
    .nunique()
    .reset_index(name='loc_count')
)
hospitals_b = loc_counts[loc_counts['loc_count'] > 1][hospital_col].unique()[:3]

show_filtered_records(hospitals_b,
    "b) Krankenhäuser, die in verschiedenen Orts-Kombinationen auftauchen")
```
